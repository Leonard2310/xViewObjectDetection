Il modello Mask R-CNN è una potente architettura per object detection e instance segmentation, sviluppata per identificare oggetti in un'immagine e, contemporaneamente, *generare una maschera binaria che definisce la forma esatta di ogni oggetto individuato*.

Mask R-CNN è un'estensione del modello Faster R-CNN. Mentre Faster R-CNN si occupa solo di:
- **Object detection**: localizzare gli oggetti tramite bounding box,
- **Classification**: determinare la classe di ogni oggetto individuato,
Mask R-CNN aggiunge un ulteriore componente:
- **Instance segmentation**: generare una maschera a livello di pixel per ogni oggetto.

Mask R-CNN si basa su tre fasi principali:

- **Backbone network per l'estrazione delle caratteristiche**

Viene usata una rete convoluzionale come ResNet o ResNeXt con un Feature Pyramid Network (FPN) per estrarre caratteristiche significative dall'immagine a diverse scale.

- **Region Proposal Network (RPN)**

La RPN è il modulo che identifica regioni candidate contenenti oggetti (chiamate Region of Interest - RoI). Questa fase genera un insieme di bounding box proposte.

- **Head per classificazione, regressione e maschere**

Ogni RoI viene raffinata e processata per:

1. Classificazione: determinare la classe dell'oggetto.
2. Regressione: affinare la bounding box.
3. Segmentazione: generare una maschera a livello di pixel usando un ramo separato (convoluzionale).

L'addestramento di Mask R-CNN richiede un dataset etichettato con:

- Bounding box per gli oggetti (come in Faster R-CNN).
- Maschere binarie per ogni istanza di ogni oggetto.
